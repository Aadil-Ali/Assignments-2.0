Q2. Scatter plot is  dot representation between x axis and y axis 

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)

plt.xlabel('Data of X Axix')
plt.ylabel("Data of Y Axis")
plt.title("Scatter plot")

```

Q3. Subplot funtion is used to create multiple plots in a single figure 

```python

a = np.array([0, 1, 2, 3, 4, 5])
b= np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,1,1)
plt.plot(a,b)

a = np.array([0, 1, 2, 3, 4, 5])
b = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,1,2)
plt.plot(a,b)

a = np.array([0, 1, 2, 3, 4, 5]) 
b = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,1,1)
plt.plot(a,b)

a = np.array([0, 1, 2, 3, 4, 5])
b = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,1,2)
plt.plot(a,b)
```

Q5. A box and whisker plotalso called a box plotdisplays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

box = [box1 , box2]

plt.boxplot(box)
```
