```python
#Q1

import seaborn as sns
import plotly.express as px
t = sns.load_dataset('titanic')

px.scatter(t,x = 'age' , y = 'fare')
```

```python
#Q2

import seaborn as sns
import plotly.express as px
tip = sns.load_dataset('tips')

px.box(tip,x = 'sex' , y = 'day')
```

```python
#Q3

px.histogram(tip , x = 'sex' , y = 'total_bill' , pattern_shape='smoker' , color='day')
```

```python
#Q5


'''
 Distplot : In Plotly Express, the `distplot` functionality is available as part of the `px.histogram` function. 
The `distplot` in Seaborn combines both a histogram and a kernel density estimation (KDE) plot, 
and in Plotly Express, you can achieve similar functionality by creating a histogram with an overlaid KDE curve.
'''
import seaborn as sns
import plotly.express as px

titanic_data = sns.load_dataset("titanic")

px.histogram(titanic_data, x="age", marginal="rug", title="Distplot of Age")


```
