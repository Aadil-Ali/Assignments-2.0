```python
import seaborn as sns
```

Q1. 
1) ScatterPlot : A scatterplot is a type of data display that shows the relationship between two numerical variables.

2) Line Plot : Draw a line plot with the possibility of several semantic groupings. The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters.


3) Histogram:
A histogram is used to represent the distribution of a single variable. Seaborn's histplot function can create histograms with various customization options.

4) Bar Plot:
A bar plot is used to compare categorical data. Seaborn's barplot function can create both vertical and horizontal bar plots.

5) Box Plot:
A box plot (box-and-whisker plot) is used to visualize the distribution of data and identify outliers. Seaborn's boxplot function can create box plots with additional features like grouping.

```python
#Q2

data = sns.load_dataset('fmri')
sns.lineplot(x = 'timepoint' , y = 'signal' , hue = 'event' , style = 'region', data = data)
```

```python
#Q3

import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')

plt.figure(figsize = (12,6))

plt.subplot(1,2,1)
sns.boxplot(x = 'pclass' , y = 'age' , data = titanic )

plt.subplot(1,2,2)
sns.boxplot(x = 'pclass' , y = 'fare' , data = titanic)
```

```python
#Q4

d = sns.load_dataset('diamonds')
sns.histplot(x = 'price' , hue = 'cut' , data = d )
```

```python
#Q5

i = sns.load_dataset('iris')
sns.pairplot(hue = 'species' , data = i)
```

```python
#Q6

import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))

sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")



```
